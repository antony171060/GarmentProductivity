```python
#Lectura del grupo de datos CSV
file_name = "garments_worker_productivity.csv"

#El método "read_csv" nos permite leer un CSV
dataset = pd.read_csv(file_name)
```

```python
#Mostrar datos y sus características
print(dataset.shape)
print(dataset.head())
print(dataset.dtypes)
print(dataset.nunique() / len(dataset))
print(dataset.isna().sum() / len(dataset))
print(dataset.to_string())
```

```python
# Si los datos son únicos con un procentaje mayor a 80% entonces eliminar la columna o característica
threshold = 0.8
unique_percentages = dataset.nunique() / len(dataset)
criteria = unique_percentages >= threshold
columns_to_filter = unique_percentages[criteria]
columns_to_filter = unique_percentages[criteria].keys()
print(columns_to_filter)
dataset.drop(columns_to_filter, axis=1, inplace=True)
#dataset.drop_duplicates(keep="first", inplace=True)
```

```python
# Mostrar estado actual de los datos
print(dataset.shape)
print(dataset.head())
print(dataset.dtypes)
print(dataset.nunique() / len(dataset))
print(dataset.isna().sum() / len(dataset))
print(dataset.to_string())
```
